```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

size_of_array = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384]

# Results in order of: merge, quick, radix, bitonic, brick (in ns)
time_merge = [4841478, 1161481, 3852243, 2403159, 3575306, 9782510, 14755149, 30243927, 72989572, 138410162, 297704061, 181440647, 685088644, 2263572335, 9223860366]
time_quick = [1390221, 254420, 1318715, 1895505, 3498664, 7913871, 12323546, 328136432, 48771122, 71905524, 216145212, 507227254, 684933385, 10011973521, 7991166032]
time_radix = [1471604, 275753, 1148839, 2315060, 2899356, 8375697, 10339548, 43249744, 40713844, 70258513, 91637656, 253275454, 1019873564, 1823370807, 46573974065]
time_bitonic = [24889, 958419, 399802, 5690464, 1719702, 2472294, 4028836, 6795451, 21771440, 56875017, 124453038, 337714301, 660483034, 3745566225, 8709981069]
time_brick = [42271, 1598419, 1970566, 3074763, 3256491, 7053822, 19518009, 50238380, 107506087, 762440484, 499847902, 3139945371, 9400779930, 42873895606, 190762916928]

# Convert from ns to ms
time_merge = [x/(10.0**6) for x in time_merge]
time_quick = [x/(10.0**6) for x in time_quick]
time_radix = [x/(10.0**6) for x in time_radix]
time_bitonic = [x/(10.0**6) for x in time_bitonic]
time_brick = [x/(10.0**6) for x in time_brick]

# Plot the results
merge_line, = plt.plot(size_of_array, time_merge, 'b--', label='Merge Sort')
radix_line, = plt.plot(size_of_array, time_radix, 'r--', label='Radix Sort')
quick_line, = plt.plot(size_of_array, time_quick, 'g--', label='Quick Sort')
bitonic_line, = plt.plot(size_of_array, time_bitonic, 'm--', label='Bitonic Sort')
brick_line, = plt.plot(size_of_array, time_brick, 'k--', label='Brick Sort')
plt.legend(handles=[merge_line, radix_line, quick_line, bitonic_line, brick_line], loc=2)
plt.title('Time VS Array Size')
plt.xlabel('Size of Array')
plt.ylabel('Time to Sort (ms)')
plt.show()

print 'Time in milliseconds of results in order: MergeSort, QuickSort, RadixSort, BitonicSort, BrickSort'
print 'With index position corresponding to array size of 2^index\n'
print time_merge
print time_quick
print time_radix
print time_bitonic
print time_brick

```
